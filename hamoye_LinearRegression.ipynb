```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df=pd.read_csv('energydata_complete.csv')
df[:5]
```

```python
df.drop(['lights'], axis=1,inplace=True)
df.drop(['date'], axis=1, inplace=True)
```

```python
scaler = MinMaxScaler()
normalized_features = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
features = normalized_features.drop(['Appliances'], axis=1)
target = normalized_features['Appliances']
```

```python
X_train,X_test,y_train,y_test=train_test_split(features, target, test_size = 0.3, random_state = 42)
```

```python
X_train[:5]
```

```python
#q12
train_x=X_train[['T2']]
train_y=X_train[['T6']] 
test_x=X_test[['T2']]
test_y=X_test[['T6']]
linearRegression=LinearRegression()
linearRegression.fit(train_x, train_y)
y_pred=linearRegression.predict(test_x)

r_square=r2_score(test_y,y_pred)
r_square=round(r_square,2)
print(r_square)
```

```python
linearRegression.fit(X_train, y_train)
y_pred=linearRegression.predict(X_test)
```

```python
#q13
mae=mean_absolute_error(y_test, y_pred)
mae=round(mae, 2)
print(mae)
```

```python
#q14
rss=np.sum((y_test-y_pred)**2)
rss=round(rss,2)
print(rss)
```

```python
#q15
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
rmse=round(rmse,3)
print(rmse)
```

```python
#q16
r_square=r2_score(y_test,y_pred)
r_square=round(r_square,2)
print(r_square)
```

```python
def get_weight(model, feat, column_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  wt=pd.DataFrame(weights).reset_index()
  wt.columns=['Features', column_name]
  wt[column_name].round(3)
  return wt

# Execute the get_weight function and store in dataframe
linear_wt = get_weight(linearRegression, X_train, 'Linear_weight')
linear_wt
```

```python
#q18
ridge_regression=Ridge(alpha=0.4)
ridge_regression.fit(X_train, y_train)
reg_pred=ridge_regression.predict(X_test)
rmse=np.sqrt(mean_squared_error(y_test, y_pred))
rmse=round(rmse,3)
print(rmse)
#same value as before
```

```python
#q19
lasso_regression=Lasso(alpha=0.001)
lasso_regression.fit(X_train, y_train)
lasso_pred=lasso_regression.predict(X_test)
lasso_wt=get_weight(lasso_regression, X_train, 'lasso_weight')
lasso_wt
```

```python
#q20
lasso_rmse=np.sqrt(mean_squared_error(y_test, lasso_pred))
lasso_rmse=round(lasso_rmse,3)
lasso_rmse
```
